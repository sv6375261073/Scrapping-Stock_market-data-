```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://economictimes.indiatimes.com/marketstats/pid-40,exchange-nse,sortby-value,sortorder-desc.cms')
```

```python
table=driver.find_element_by_class_name('dataContainer')
```

```python
text=table.text
```

```python
trades=[i for i in text.replace('Nifty','@Nifty').split('@') if len(i)>0]
```

```python
trades=[line.split('\n')[:5] for line in trades]
```

```python
trade_names=['Name','Value','Prev_Close','Change','%Change']
trade_chart={line[0]:{col:val for col,val in zip(trade_names,line)} for line in trades}
```
